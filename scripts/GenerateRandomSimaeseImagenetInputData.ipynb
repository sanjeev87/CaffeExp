```python
import numpy as np
file = open('SandyData/allWordsSorted.txt', 'r')
#for str in file:
#    print str
ad = 'advertisement'
rep = 'representative' 
well = 'well'
go = "go"
ext = '.png'
path = "/raid/swiraatm/words_data/images/"

numSamples = 3000

Words = [ad,go,well]

file_train = open("train.txt", "w")
file_train_p = open("train_p.txt", "w")

for i in range(numSamples):
    rand1 = np.random.randint(3)
    rand2 = np.random.randint(3)
    randImg1 = np.random.randint(299)+1
    randImg2 = np.random.randint(299)+1
    
    similar = "0"
    if rand1 == rand2:
        similar = "1"
    
    string_train = path + Words[rand1] + "%d" % randImg1 + ".png"+ " " + similar + "\n"
    string_train_p = path + Words[rand2] + "%d" % randImg2 + ".png"+ " " + similar + "\n"
    
    file_train.write(string_train)
    file_train_p.write(string_train_p)

file_train.close()
file_train_p.close()
```
